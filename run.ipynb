```python
import random
from tqdm import tqdm
```

```python
#Sample for 1V1

time_limit = None
warning_limit = 3
seed = random.randint(0,1000000)

displayer = None
#displayer = TextGameDisplayer()

replays = []
count = [0]*len(players)
total = 100

for _ in tqdm(range(total)):
    players = [NaivePlayer(0), RandomPlayer(1)]
    gr = AdvanceGameRunner(players,
                           seed=seed,
                           time_limit=time_limit,
                           warning_limit=warning_limit,
                           displayer=displayer)
    replay = gr.Run()
    replays.append(replay)
        
    if replay[0][0]<replay[1][0]:
        count[1]+=1
    else:
        count[0]+=1
        

print ()
for i,p in enumerate(players):
    print (p.__class__.__name__,count[i])
print (total)
```
