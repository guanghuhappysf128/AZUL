```python
from advance_model import AdvanceGameRunner, ReplayRunner
from naive_player import NaivePlayer
from thinking_player import ThinkingPlayer
from displayer import TextGameDisplayer
from utils import *
```

```python
#Sample Game Run

players = [NaivePlayer(0), NaivePlayer(1), NaivePlayer(2),NaivePlayer(3)]

gr = AdvanceGameRunner(players,
                       seed=1,
                       time_limit=1.0,
                       warning_limit=3,
                       displayer=TextGameDisplayer())
replay = gr.Run()
```

```python
for i in range(len(players)):
    print ("Score for Player {}: {}".format(i,replay[i][0]))
```

```python
#Sample Replay Run

rr = ReplayRunner(replay,
                  displayer = TextGameDisplayer())
rr.Run()
```
